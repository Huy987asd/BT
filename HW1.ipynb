```python
from skfuzzy.control.state import StatePerSimulation
from skfuzzy.control.controlsystem import ControlSystem
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl 
time = ctrl.Antecedent(np.arange(0,181,1),'time')
rice = ctrl.Antecedent(np.arange(100,2001,1),'rice')
power = ctrl.Antecedent(np.arange(0,101,1),'power')
time['Very Fast'] = fuzz.trimf(time.universe,[0,15,20])
time['Fast'] = fuzz.trimf(time.universe,[20,25,35])
time['Normal'] = fuzz.trimf(time.universe,[25,35,45])
time['Slow'] = fuzz.trimf(time.universe,[45,60,70])
time['Very Slow'] = fuzz.trimf(time.universe,[70,120,180])
rice['VL'] = fuzz.trimf(rice.universe,[100,250,350])
rice['L'] = fuzz.trimf(rice.universe,[350,550,700])
rice['Normal'] = fuzz.trimf(rice.universe,[700,850,950])
rice['M'] = fuzz.trimf(rice.universe,[950,1200,1500])
rice['VM'] = fuzz.trimf(rice.universe,[1500,1800,2000])
power['verylow'] = fuzz.trimf(power.universe,[0,20,30])
power['low'] = fuzz.trimf(power.universe,[20,40,50])
power['medium'] = fuzz.trimf(power.universe,[40,50,70])
power['high'] = fuzz.trimf(power.universe,[60,70,90])
power['veryhigh'] = fuzz.trimf(power.universe,[80,100,100])
time.view()
rice.view()
power.view()

rule1 = ctrl.Rule(time['Very Fast']& rice['VL'], power['verylow'])
rule2 = ctrl.Rule(time['Fast']& rice['VL'], power['verylow'])
rule3 = ctrl.Rule(time['Normal']& rice['VL'], power['low'])
rule4 = ctrl.Rule(time['Slow']& rice['VL'], power['low'])
rule5 = ctrl.Rule(time['Very Slow']& rice['VL'], power['medium'])
rule6 = ctrl.Rule(time['Very Fast']& rice['L'], power['verylow'])
rule7 = ctrl.Rule(time['Fast']& rice['L'], power['low'])
rule8 = ctrl.Rule(time['Normal']& rice['L'], power['low'])
rule9 = ctrl.Rule(time['Slow']& rice['L'], power['medium'])
rule10 = ctrl.Rule(time['Very Slow']& rice['L'], power['medium'])
rule11 = ctrl.Rule(time['Very Fast']& rice['Normal'], power['low'])
rule12 = ctrl.Rule(time['Fast']& rice['Normal'], power['medium'])
rule13 = ctrl.Rule(time['Normal']& rice['Normal'], power['medium'])
rule14 = ctrl.Rule(time['Slow']& rice['Normal'], power['high'])
rule15 = ctrl.Rule(time['Very Slow']& rice['Normal'], power['high'])
rule16 = ctrl.Rule(time['Very Fast']& rice['M'], power['medium'])
rule17 = ctrl.Rule(time['Fast']& rice['M'], power['medium'])
rule18 = ctrl.Rule(time['Normal']& rice['M'], power['high'])
rule19 = ctrl.Rule(time['Slow']& rice['M'], power['veryhigh'])
rule20 = ctrl.Rule(time['Very Slow']& rice['M'], power['veryhigh'])
rule21 = ctrl.Rule(time['Very Fast']& rice['VM'], power['high'])
rule22 = ctrl.Rule(time['Fast']& rice['VM'], power['high'])
rule23 = ctrl.Rule(time['Normal']& rice['VM'], power['veryhigh'])
rule24 = ctrl.Rule(time['Slow']& rice['VM'], power['veryhigh'])
rule25 = ctrl.Rule(time['Very Slow']& rice['VM'], power['veryhigh'])


powering_ctrl = ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10,rule11,rule12,rule13,rule14,rule15,rule16,rule17,rule18,rule19,rule20,rule21,rule22,rule23,rule24,rule25])
powering = ctrl.ControlSystemSimulation(powering_ctrl)
powering.input['time'] = 10
powering.input['rice'] = 9


powering.compute()
print(powering.output['power'])
power.view(sim=powering)


```
